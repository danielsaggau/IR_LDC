```python
tokenizer = AutoTokenizer.from_pretrained('bert-base-cased')
dataset = dataset.map(lambda e: tokenizer(e['facts'], truncation=True, padding='max_length'), batched=True)
```

```python
dataset.set_format(type='torch', columns=['labels', 'silver_rationales', 'gold_rationales'])
dataloader = torch.utils.data.DataLoader(dataset, batch_size=32)
next(iter(dataloader))
#dataloader
```
