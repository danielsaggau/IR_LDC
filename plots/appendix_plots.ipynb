```python
import pandas as pd
df_time = pd.read_excel(r'/content/performance_benchmark.xlsx', sheet_name='time')
df_sigma = pd.read_excel(r'/content/performance_benchmark.xlsx', sheet_name='sigma')
df_subnetworks = pd.read_excel(r'/content/performance_benchmark.xlsx', sheet_name='subnetworks')
print(df_sigma)
```

```python

plt.rcParams.update(plt.rcParamsDefault)
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(nrows=1,ncols= 2,sharex=False, sharey=False, figsize=(10,6))
plt.rcParams['axes.grid'] = True
plt.rcParams.update({'font.size': 10})
plt.rcParams.update({'axes.titlesize': 'large'}) 
sns.lineplot(data=df_sigma, x='sigma', y='MacroF1',ax=ax[0],marker="o", color='black')
sns.lineplot(data=df_sigma, x='sigma', y='MicroF1',ax =ax[1],marker="o", color='black')
plt.grid()

ax[1].set_ylabel('Micro F1 Performance')
ax[0].set_ylabel('Macro F1 Performance')
ax[1].set_xlabel('$\sigma$')
ax[0].set_xlabel('$\sigma$')
ax[0].set_title('(a) Macro-F1', size=12)
ax[1].set_title('(b) Micro-F1', size=12)


ax[0].set_ylim([60, 80])
ax[1].set_ylim([60, 80])

for spine in ['top', 'right']:
    ax[0].spines[spine].set_visible(False)
for spine in ['top', 'right']:
    ax[1].spines[spine].set_visible(False)
plt.grid()
fig.tight_layout(pad=.8)
plt.savefig('sigma.png', dpi=500)
```
